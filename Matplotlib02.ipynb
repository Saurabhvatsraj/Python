```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# pie chart--- percentage distribuation

sales=[190,200,105,40]
mylabels=['IND','AUS','SL','PAK']
plt.xlabel('COUNTRY')
plt.ylabel('SALES')
plt.pie(sales,labels=mylabels)
```

```python
sales=[190,200,105,40]
mylabels=['IND','AUS','SL','PAK']
plt.xlabel('COUNTRY')
plt.ylabel('SALES')
plt.pie(sales,labels=mylabels,autopct='1')
```

```python
sales=[190,200,105,40]
mylabels=['IND','AUS','SL','PAK']
plt.xlabel('COUNTRY')
plt.ylabel('SALES')
plt.pie(sales,labels=mylabels,autopct='%.2f')
```

```python
# subplot-----to see multiple graphs
x=[23,56,45,87]
y=[90,78,34,76]
plt.subplot(2,2,1)
plt.plot(x,y)

x=[1,2,3,4]
y=[50,60,30,10]
plt.subplot(1,2,2)
plt.plot(x,y)
```
